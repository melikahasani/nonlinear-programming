```python
import numpy as np
import numpy.linalg as la
def pure_netwon(f, g, h, x0, epsilon):
  x = x0
  gval = g(x)
  hval = h(x)
  iter = 0
  while la.norm(gval) > epsilon and iter < 10000:
    iter = iter + 1
    x = x - la.solve(hval, gval) #hval*x = gval ----> x = inverse(hval) * gval ----> x = d = inverse of hessian matrix of g * gradient of g
    fun_val = f(x)
    gval = g(x)
    hval = h(x)
    #print('iter = {0:3d}'.format(iter),\
     # 'f(x) = {0:2.6f}'.format(fun_val))
    if (iter==10000):
      print('did not converge')
  return x, iter
```

```python
import numpy as np
import numpy.linalg as la
f = lambda x : x[0]**2 + 0.01 * x[1]**2
g = lambda x : np.array([2 * x[0], 0.02 * x[1]])
x0 = np.array([0.01, 1.0])
epsilon = 1e-6
h = lambda x : np.diag([2, 0.02])
x, iter = pure_netwon(f, g, h, x0, epsilon)
print(x, iter )
```
